```python
import numpy as np
import matplotlib.pyplot as plt

data = {k: 0 for k in range(0, 7500)}

# read communities from file and count their number of entries
with open('ig_communities_output_120h.metis', 'r') as inputFile:
    for line in inputFile:
        splittedLine = line.split(' ')
        data[len(splittedLine)] += 1

print(data)

def count_distinct_numbers(filename):
    numbers_set = set()
    with open(filename, 'r') as file:
        for line in file:
            # Split the line into numbers using whitespace as a separator
            numbers = line.split()
            # Add each number to the set (duplicates will automatically be ignored)
            numbers_set.update(numbers)
    return len(numbers_set)

# Replace 'filename.txt' with the path to your file
number_of_distinct_numbers = count_distinct_numbers('ig_communities_input_backup')
print("Total number of distinct numbers:", number_of_distinct_numbers)
```

```python
fig = plt.figure(figsize = (10, 5))

plt.stairs(data.values(),  range(0, 7501))
plt.yscale('log')
plt.xscale('log')

plt.ylabel("# of communities")
plt.xlabel("# of members in community")

plt.show()
```
